```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
prenoms = pd.read_csv("dpt2018.csv",sep=";")
prenoms
```

```python
top20=prenoms.groupby(prenoms.preusuel).nombre.sum().sort_values()[-20:]
top20
```

```python
import seaborn as sns
sns.barplot(y=top20.index, x = top20.values).set_title("Volume des 20 prénoms les plus donnés entre 1900 et 2018")
```
