# Firstname diversity study

```python
nat = pd.read_csv("nat2018.csv",sep=';')
nat
```

```python
#Afficher image
#![groupby - apply - combine](Groupby-Apply-Combine.jpeg)
```

```python
nat=nat[nat.annais!='XXXX']
nat
```

```python
#PROPORTION des naissances avec un prénom par rapport à la somme des naissances pour cette année
def add_prop(group):
    group['prop'] = group.nombre/group.nombre.sum()
    return group

nat=nat.groupby(['annais','sexe']).apply(add_prop)

nat.sort_values(by='prop')
```

```python
#SANITY CHECK I :
nat.groupby(["annais","sexe"]).prop.sum()
```

```python
#SANITY CHECK II :
sums1900=nat[nat.annais=='1900'].groupby(['sexe']).nombre.sum()
sums1900
```

```python
#SANITY CHECK II bis (auto) : AUTOMATIC, FACTORISED

#année du prenom avec la plus grande proportion
YrMaxProp=nat.sort_values(by='prop')[-1:].annais
YrMaxProp
```

```python
#ILOC

#PERMET D ACCEDER AU CONTENU D UNE CELLULE 

YrMaxProp=nat.sort_values(by='prop')[-1:].annais.iloc[0]
#YrMaxProp=int(YrMaxProp)
YrMaxProp
```

```python
#nb de naissances du genre et de l'année du prenom qui a eu la plus forte proportion

birthMaxProp=nat.sort_values(by='prop')[-1:].nombre.iloc[0]
birthMaxProp
```

```python
#max proportion
maxprop=nat.sort_values(by='prop')[-1:].prop.iloc[0]
maxprop
```

```python
sumsYrMaxProp=nat[nat.annais==YrMaxProp].groupby(["annais","sexe"]).nombre.sum()
sumsYrMaxProp
```

```python
sumsYrMaxProp=nat[nat.annais==YrMaxProp].groupby(["annais","sexe"]).nombre.sum().iloc[1]
sumsYrMaxProp
```

```python
if birthMaxProp / sumsYrMaxProp == maxprop:
    print("Sanity Check 2 ... OK")
```

```python
def get_top5(df,column='nombre'):
    return df.sort_values(by=column)[-5:]

get_top5(nat)
```

```python
def get_top1000(df,column='nombre'):
    return df.sort_values(by=column)[-1000:]

get_top1000(nat).sort_values(by='prop')
```

```python
nat.groupby(["annais",'sexe']).apply(get_top5)
```
